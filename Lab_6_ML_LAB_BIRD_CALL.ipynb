```python
#A2
import numpy as np
import matplotlib.pyplot as plt

# Step activation function
def step_activation(x):
    return 1 if x >= 0 else 0

# Bi-Polar Step function
def bipolar_step_activation(x):
    return 1 if x >= 0 else -1

# Sigmoid function
def sigmoid_activation(x):
    return 1 / (1 + np.exp(-x))

# ReLU function
def relu_activation(x):
    return max(0, x)

# Perceptron class
class Perceptron:
    def __init__(self, activation='step', learning_rate=0.05, weights=None):
        self.learning_rate = learning_rate
        self.activation = activation
        self.weights = weights if weights is not None else np.array([-10, 0.2, -0.75])

    def predict(self, inputs):
        weighted_sum = np.dot(self.weights[1:], inputs) + self.weights[0]

        if self.activation == 'step':
            return step_activation(weighted_sum)
        elif self.activation == 'bipolar_step':
            return bipolar_step_activation(weighted_sum)
        elif self.activation == 'sigmoid':
            return sigmoid_activation(weighted_sum)
        elif self.activation == 'relu':
            return relu_activation(weighted_sum)

    def train(self, training_inputs, labels, max_epochs=1000, error_threshold=0.002):
        epoch_errors = []
        for epoch in range(max_epochs):
            total_error = 0
            for inputs, label in zip(training_inputs, labels):
                prediction = self.predict(inputs)

                # Error calculation for bipolar step and sigmoid
                if self.activation == 'bipolar_step':
                    error = label - prediction
                else:
                    error = label - prediction

                total_error += error ** 2

                # Update weights
                self.weights[1:] += self.learning_rate * error * inputs
                self.weights[0] += self.learning_rate * error

            epoch_errors.append(total_error)

            if total_error <= error_threshold:
                print(f'Converged with {self.activation} activation after {epoch + 1} epochs.')
                break
        else:
            print(f'Reached maximum epochs without convergence for {self.activation} activation.')

        return epoch_errors

# AND gate training data
training_inputs = np.array([[0, 0],
                            [0, 1],
                            [1, 0],
                            [1, 1]])

labels = np.array([0, 0, 0, 1])

# Activation functions to compare
activations = ['step', 'bipolar_step', 'sigmoid', 'relu']
initial_weights = np.array([-10, 0.2, -0.75])

# Store error histories for comparison
error_histories = {}
epochs_required = {}

# Train the perceptron with each activation function and collect errors
for activation in activations:
    perceptron = Perceptron(activation=activation, weights=initial_weights.copy())
    error_values = perceptron.train(training_inputs, labels)
    error_histories[activation] = error_values
    epochs_required[activation] = len(error_values)

# Plotting the error values for each activation function
plt.figure(figsize=(10, 6))

for activation, errors in error_histories.items():
    plt.plot(errors, label=f'{activation} activation')

plt.title('Epochs vs. Sum-Squared Error for Different Activation Functions')
plt.xlabel('Epochs')
plt.ylabel('Sum-Squared Error')
plt.yscale('log')
plt.grid(True)
plt.legend()
plt.show()

# Print out the epochs required for each activation function
print("Epochs required for each activation function to converge:")
for activation, epochs in epochs_required.items():
    print(f"{activation}: {epochs} epochs")
```

```python
#A6
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid activation function
def sigmoid_activation(x):
    return 1 / (1 + np.exp(-x))

# Perceptron class with sigmoid activation
class Perceptron:
    def __init__(self, learning_rate=0.05, weights=None):
        self.learning_rate = learning_rate
        # Initialize weights (4 inputs + 1 bias)
        self.weights = weights if weights is not None else np.random.randn(5)

    def predict(self, inputs):
        # Weighted sum (w1*x1 + w2*x2 + ... + bias)
        weighted_sum = np.dot(self.weights[1:], inputs) + self.weights[0]
        # Apply sigmoid activation function
        return sigmoid_activation(weighted_sum)

    def train(self, training_inputs, labels, max_epochs=1000, error_threshold=0.002):
        epoch_errors = []
        for epoch in range(max_epochs):
            total_error = 0
            for inputs, label in zip(training_inputs, labels):
                # Prediction
                prediction = self.predict(inputs)
                # Error calculation (label - prediction)
                error = label - prediction
                total_error += error ** 2

                # Update weights using gradient descent
                # Adjust based on sigmoid derivative
                self.weights[1:] += self.learning_rate * error * prediction * (1 - prediction) * inputs
                self.weights[0] += self.learning_rate * error * prediction * (1 - prediction)  # Bias update

            epoch_errors.append(total_error)
            # Stop if error threshold is met
            if total_error <= error_threshold:
                print(f"Converged after {epoch + 1} epochs.")
                break
        else:
            print("Did not converge within the maximum epochs.")

        return epoch_errors

# Customer data (features: candies, mangoes, milk packets, payment)
training_inputs = np.array([
    [20, 6, 2, 386],
    [16, 3, 6, 289],
    [27, 6, 2, 393],
    [19, 1, 2, 110],
    [24, 4, 2, 280],
    [22, 1, 5, 167],
    [15, 4, 2, 271],
    [18, 4, 2, 274],
    [21, 1, 4, 148],
    [16, 2, 4, 198]
])

# Labels (High value transaction? Yes = 1, No = 0)
labels = np.array([1, 1, 1, 0, 1, 0, 1, 1, 0, 0])

# Initialize perceptron with random weights
initial_weights = np.random.randn(5)  # 4 input features + 1 bias
learning_rate = 0.05
perceptron = Perceptron(learning_rate=learning_rate, weights=initial_weights)

# Train the perceptron
error_values = perceptron.train(training_inputs, labels)

# Plotting the error values over epochs
plt.plot(error_values)
plt.title('Epochs vs. Sum-Squared Error')
plt.xlabel('Epochs')
plt.ylabel('Sum-Squared Error')
plt.yscale('log')
plt.grid(True)
plt.show()

# Test the perceptron with a new customer (example)
new_customer = np.array([25, 5, 3, 300])  # Candies, Mangoes, Milk Packets, Payment
prediction = perceptron.predict(new_customer)
print(f"Prediction for new customer (high value transaction): {'Yes' if prediction >= 0.5 else 'No'}")
```

```python
#A8
import numpy as np

# Sigmoid activation function and its derivative
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

# Input and output for AND gate
inputs = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
outputs = np.array([[0], [0], [0], [1]])  # Expected outputs for AND gate

# Initialize weights randomly
np.random.seed(1)
v = np.random.rand(2, 2)  # Weights for input to hidden layer
w = np.random.rand(2, 1)  # Weights for hidden to output layer

# Learning rate and parameters
alpha = 0.05  # Learning rate
max_iterations = 1000  # Maximum iterations for learning
target_error = 0.002  # Error threshold for convergence

# Training the neural network using backpropagation
for iteration in range(max_iterations):
    # Forward propagation
    # Calculate hidden layer outputs (H1 and H2)
    hidden_layer_output = sigmoid(np.dot(inputs, v))

    # Calculate final output (O1)
    output_layer_output = sigmoid(np.dot(hidden_layer_output, w))

    # Calculate the error
    error = outputs - output_layer_output

    # Print the error every 100 iterations to track progress
    if iteration % 100 == 0:
        print(f"Iteration {iteration} - Error: {np.mean(np.abs(error))}")

    # Check if the error has converged
    if np.mean(np.abs(error)) <= target_error:
        print(f"Converged at iteration {iteration} with error {np.mean(np.abs(error))}")
        break

    # Backpropagation
    # Calculate the delta for output layer
    output_delta = error * sigmoid_derivative(output_layer_output)

    # Calculate the delta for hidden layer
    hidden_delta = output_delta.dot(w.T) * sigmoid_derivative(hidden_layer_output)

    # Update weights
    w += alpha * hidden_layer_output.T.dot(output_delta)
    v += alpha * inputs.T.dot(hidden_delta)

#Final weights and outputs after training
print("Final weights from input to hidden layer (v):")
print(v)
print("Final weights from hidden to output layer (w):")
print(w)

#Testing the trained network
print("\nTesting the trained network:")
for i in range(4):
    hidden_layer_output = sigmoid(np.dot(inputs[i], v))
    output = sigmoid(np.dot(hidden_layer_output, w))
    print(f"Input: {inputs[i]} - Predicted Output: {output[0]:.4f} - Expected Output: {outputs[i][0]}")

```

```python
#A11
import numpy as np
from sklearn.neural_network import MLPClassifier
import matplotlib.pyplot as plt

def plot_decision_boundary(clf, X, y, title):
    # Create a mesh grid for plotting decision boundaries
    xx, yy = np.meshgrid(np.linspace(-0.5, 1.5, 100), np.linspace(-0.5, 1.5, 100))
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # Plot decision boundary
    plt.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.Paired)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k', marker='o', cmap=plt.cm.Paired)
    plt.title(title)
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.grid(True)

# Define AND gate training data
X_and = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])
y_and = np.array([0, 0, 0, 1])

# Define XOR gate training data
X_xor = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])
y_xor = np.array([0, 1, 1, 0])

# Initialize the MLPClassifier for AND gate
mlp_and = MLPClassifier(hidden_layer_sizes=(2,), activation='relu', solver='adam', max_iter=1000, random_state=42)
mlp_and.fit(X_and, y_and)
y_pred_and = mlp_and.predict(X_and)

print("AND Gate Results")
print("Predictions:", y_pred_and)
print("Actual Labels:", y_and)

# Initialize the MLPClassifier for XOR gate
mlp_xor = MLPClassifier(hidden_layer_sizes=(2,), activation='relu', solver='adam', max_iter=1000, random_state=42)
mlp_xor.fit(X_xor, y_xor)
y_pred_xor = mlp_xor.predict(X_xor)

print("\nXOR Gate Results")
print("Predictions:", y_pred_xor)
print("Actual Labels:", y_xor)

# Plot decision boundaries
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plot_decision_boundary(mlp_and, X_and, y_and, 'AND Gate MLP Decision Boundary')

plt.subplot(1, 2, 2)
plot_decision_boundary(mlp_xor, X_xor, y_xor, 'XOR Gate MLP Decision Boundary')

plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt

def load_data(file_path):
    # Load dataset from an Excel file
    data = pd.read_excel(file_path)
    print(data.head())
    return data

def preprocess_data(data):
    # Prepare features and labels for training
    X = data.drop(['species'], axis=1).values  # Features
    y = data['species'].values  # Labels

    # Encode labels as numbers
    label_encoder = LabelEncoder()
    y = label_encoder.fit_transform(y)

    print(f"Feature shape: {X.shape}")
    print(f"Labels shape: {y.shape}")
    return X, y

def split_and_scale_data(X, y):
    # Split data into training and testing sets, and scale the features
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    print(f"Training data shape: {X_train_scaled.shape}")
    print(f"Testing data shape: {X_test_scaled.shape}")
    return X_train_scaled, X_test_scaled, y_train, y_test

def train_mlp_classifier(X_train, y_train):
    # Initialize and train the MLPClassifier
    mlp = MLPClassifier(hidden_layer_sizes=(10, 10), activation='relu', solver='adam', max_iter=2000, random_state=42)
    mlp.fit(X_train, y_train)
    print("MLP Classifier trained.")
    return mlp

def evaluate_model(mlp, X_test, y_test):
    # Evaluate the model and print results
    y_pred = mlp.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"\nAccuracy: {accuracy:.2f}")
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))
    return y_pred

def plot_decision_boundary(clf, X, y, title):
    # Plot decision boundaries (only applicable for 2D feature spaces)
    if X.shape[1] != 2:
        print("Decision boundary plotting is only possible with exactly 2 features.")
        return

    xx, yy = np.meshgrid(np.linspace(X[:, 0].min() - 1, X[:, 0].max() + 1, 100),
                         np.linspace(X[:, 1].min() - 1, X[:, 1].max() + 1, 100))
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.Paired)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k', marker='o', cmap=plt.cm.Paired)
    plt.title(title)
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.grid(True)


file_path = "C:\\Users\\Administrator\\Downloads\\bird_species_features (1).xlsx"

# Load data
data = load_data(file_path)

# Preprocess data
X, y = preprocess_data(data)

# Split and scale data
X_train_scaled, X_test_scaled, y_train, y_test = split_and_scale_data(X, y)

# Train MLPClassifier
mlp = train_mlp_classifier(X_train_scaled, y_train)

# Evaluate model
evaluate_model(mlp, X_test_scaled, y_test)

# Plot decision boundary if applicable (only if there are exactly 2 features)
if X.shape[1] == 2:
    plt.figure(figsize=(8, 6))
    plot_decision_boundary(mlp, X_test_scaled, y_test, 'MLP Decision Boundary')
    plt.show()
```
