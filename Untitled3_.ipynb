```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def get_dataset(path):
  df= pd.read_excel("/content/bird_species_features.xlsx") # loading the dataset
  return df

def histogram(data, feature, bins=10):# plotting histogram for a feature in our dataset

    plt.hist(data[feature], bins=bins, edgecolor='black')
    plt.title(f'Histogram of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()

def meanVariance(data, feature):
    #Calculates and returns the mean and variance of a feature
    mean_value = data[feature].mean()
    variance_value = data[feature].var()
    return mean_value, variance_value

path='/content/bird_species_features.xlsx'

df1=get_dataset(path)
feature='feature_0'
histogram(df1, feature, bins=10)# bins are the  no. of columns
mean, variance = meanVariance(df1, feature)
print(f"Mean of {feature}: {mean}")
print(f"Variance of {feature}: {variance}")


```
