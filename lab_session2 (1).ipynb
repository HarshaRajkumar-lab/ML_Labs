```python
import pandas as pd
import numpy as np
def read_excel(excel_path):
    dataframe1 = pd.read_excel(excel_path, sheet_name ="Purchase data")
    return dataframe1
def get_mat_A(dataframe):
  return dataframe[['Candies (#)','Mangoes (Kg)','Milk Packets (#)']].to_numpy()
def get_mat_C(dataframe):
  return dataframe[['Payment (Rs)']].to_numpy()
def find_rank(mat):
  return np.linalg.matrix_rank(mat)
def pseudo_inv(mat):
  return np.linalg.pinv(mat)
def find_X(inv_A,C):
  return np.dot(inv_A,C)
def bin(mat):
  mat_new =[]
  for i in mat:
    if i >= 200:
      mat_new.append(1)
    else:
      mat_new.append(0)
  return mat_new
def calc_c(arr,Y):
  val = np.dot(arr,Y)
  #print(val)
  if round(val) ==1:
    print("RICH")
  elif round(val) ==0 :
    print("POOR")
  else:
    print("NONE")


path = '/content/drive/MyDrive/Colab Notebooks/Lab Session Data.xlsx'
df1 = read_excel(path)
A = get_mat_A(df1)
#print(A)
C = get_mat_C(df1)
#print(C)
dimensionality = A.shape[1]
print("dimensionality ",dimensionality)
no_vectors = A.shape[0]
print("Number of vectors: ", no_vectors)
rank_A = find_rank(A)
print("Rank of A: ",rank_A)
A_inv = pseudo_inv(A)
print("Pseudo inverse of A:")
print(A_inv)
X = find_X(A_inv,C)
print("Cost: ")
print(X)
C_new = bin(C)
Y = find_X(A_inv,C_new)
print("new X matrix: ")
print(Y)
for i in range (no_vectors):
  calc_c(A[i],Y)

```

```python
import pandas as pd
import numpy as np
import statistics
def get_price_list(df1):
  list1 =[]
  for i in range(df1.shape[0]):
    list1.append(df1.iloc[i,3])
  return list1
def read_excel(excel_path):
    dataframe1 = pd.read_excel(excel_path, sheet_name ="IRCTC Stock Price")
    return dataframe1

path = "/content/drive/MyDrive/Colab Notebooks/Lab Session Data.xlsx"
df1 = read_excel(path)
price_list = get_price_list(df1)
mean_price= statistics.mean(price_list)
print("Mean = ", mean_price)
variance_price = statistics.variance(price_list)
print("Variance = ", variance_price)
#mean price for wednesdays
list2 =[]
for i in range(df1.shape[0]):
  if df1.iloc[i, 2] == "Wed":
    list2.append(df1.iloc[i,3])
mean_price_wed = statistics.mean(list2)
print("Mean price on Wednesdays' = ", mean_price_wed)
#comparison
if mean_price_wed > mean_price:
  print("Mean on wednesdays is higher than that of total mean")
elif mean_price > mean_price_wed:
  print("Mean on wednesdays is lower than that of total mean")
else:
  print("Mean on wednesdays is equals that of total mean")
#mean price for Apr
list3 =[]
for i in range(df1.shape[0]):
  if df1.iloc[i, 1] == "Apr":
    list3.append(df1.iloc[i,3])
print(list3)
mean_price_apr = statistics.mean(list3)
#comparison
if mean_price_apr > mean_price:
  print("Mean on April is higher than that of total mean")
elif mean_price > mean_price_apr:
  print("Mean on April is lower than that of total mean")
else:
  print("Mean on April is equals that of total mean")

```
